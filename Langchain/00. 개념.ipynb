## 랭체인의 개념 
출처 : 모두의 AI
- 언어모델로 구동되는 애플리케이션을 개발하기 위한 프레임워크 
- 언어모델을 잘 활용할수 있도록 하는 도구 

### 쳇GPT 한계
- 정보 접근 제한 : 최신정보를 물어볼때 답변이 없음  -> 벡터스토어 또는 에이전트 활용 검색
- 토큰 제한 -> 텍스트스플리터를 통한 해결 
- 환각현상 : 정보가 없는데 거짓말을 한다. -> 주어진 문서에 대한 답변만 하도록 정의 

### Chat GPT 개량 
- Fine-turning
- N-shot Learning
- In-contact Learning : 세분화된 전문 영역 학습 ( 랭체인 역활 )

### 랭체인 구조 
- LLM : 초거대 언어모델 (사용제한없음)
- 프롬프트 : 언어모델에게 지시하는 명령문 
- 인덱스 : 쉽게 탐색할수 있도록 구조화하는 모듈
- 메모리 : 채팅이력을 기록
- 체인 : LLM 사슬을 형성 (프롬프트의 연결) 
- 에이전트 : 기존 프롬프트 템플릿으로 수행할수 없는 작업을 가능케 하는 모듈 

### PDF 챗봇 구축
아래 모든 과정에 LangChain 이 관여를 합니다.  
1. 문서업로드 
2. 문서 분할 : 여러문서로 쪼개기 
3. 문서 임베딩 : 문서의 수치화
4. 임베딩 검색 : 질문과 연관성이 높은 문서 추출 
5. 답변 생성 (QA Chain)

## 초거대 언어모델
트렌스포머 아키텍처를 기반으로 하여 모델의 용도에 따라 트랜스포머의 Encoder, Decoder를 개별 또는 통합 사용하는 추세
